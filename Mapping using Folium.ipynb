```python
import folium

#Regular map

canada_map= folium.Map(
location=[56.130,-106.35],
zoom_start=4)

canada_map
```

```python
#Stamen Toner - For Rivers and Coastal zones

world_map= folium.Map(
location=[56.130,-106.35],
zoom_start=4,
tiles='Stamen Toner')

world_map
```

```python
#Stamen Terrain - For Vegetation
world_map= folium.Map(
location=[56.130,-106.35],
zoom_start=4,
tiles='Stamen Terrain')

world_map
```

```python
#adding marker

ontario= folium.map.FeatureGroup()

ontario.add_child(
    folium.CircleMarker(
        [51.25,-85.32],
        radius=5,
        color='red',
        fill_color= 'Red'))
    
canada_map.add_child(ontario)

#Adding popup
folium.Marker([51.25,-85.32],
             popup='Ontario').add_to(canada_map)
canada_map

```
